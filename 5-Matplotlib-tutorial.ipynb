# MatplotLib Tutorial

```python
import matplotlib.pyplot as plt

%matplotlib inline  # writing this line so that plot.show() is not required everytime to plot a graph
```

```python
import numpy as np
```

```python
## Simple Examples

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## plotting using matplotlib

# plt scatter

plt.scatter(x,y,c='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y = x**2
```

```python
# plt plot

plt.plot(x,y,'g*', linestyle='dashed', linewidth=4, markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*-')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')

plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
x = np.arange(1,11)
y = 3*x+5

plt.title('Matplotlob Demo')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
```

```python
np.pi
```

```python
pi = np.pi

x = np.arange(0, 4*pi, 0.1)
y = np.sin(x)
plt.title('Sin waveform')

plt.plot(x,y)
```

```python
x = np.arange(0, 5*pi , 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin, 'r--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos, 'g')
plt.title('cos')
```

```python
## Bar Plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color='r')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])

plt.hist(a, bins=20) # default bin number is 10
plt.title('Histogarm')
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# reactangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

## Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0) #explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
