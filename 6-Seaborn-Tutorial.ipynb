# Seaborn Tutorial

## Disrtribution Plots

- distplot
- jointplot --> For 2 features
- pairplot --> For more than 2 features

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete diensions or every types. It is very important in Feature Selection

```python
df.corr() # to get pearson correlation coefficient
```

```python
sns.heatmap(df.corr())
```

## Jointplot

A joinplot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatter plot, a hexbin plot, a 2D Histogram or a 2D density plot.

## Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

## Pair plot

A 'pair plot' is also known as scatterplot, in which ome variable in the same data is matched with another variable's value, like this: Pair plots are just elaborations on this, showing all variables paired with all other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10) # kde --> Kernel Density Estimation
```

## Categorical Plots

Seaborn helps us in doing analysis on Categorical data points.

- boxplot
- violinplot
- countplot
- bar plot


```python
## Count plot

print(df['sex'].value_counts())

sns.countplot('sex', data=df)
```

```python
print(df['day'].value_counts())

sns.countplot('day', data=df)
```

```python
print(df['smoker'].value_counts())

sns.countplot(y='smoker', data=df)
```

```python
## Bar Plot

sns.barplot(x='total_bill', y='sex', data=df)  # specifying both x and y is necessary
```

```python
sns.barplot(x='smoker', y='total_bill', data=df)
```

### Box Plot

A box and whisker plot (sometimes called BOX PLOT) is a graph that presents from a five-number summary.

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot('day','total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
# Categorize my data based on some other categories

sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

### Violin Plot

Violin plot helps us to see both the distribution in terms of Kernel Density Estimation (kde) and the box plot.

```python
sns.violinplot(x='day', y='total_bill', data=df, palette='rainbow')
```
